## Import Dependencies

```python
import numpy as np
from collections import Counter
```

## Euclidean Distance

d = √[(x2 - x1)2 + (y2 - y1)2]



```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))
```

### Euclidean Step by Step

```python
point1 = [4, 5]
point2 = [1, 3]

print(f'point 1 - point 2 = {point1} - {point2} = {np.array(point1) - np.array(point2)}')
print(f'(point 1 - point 2)**2 = {(np.array(point1) - np.array(point2))**2}')
print(f'sum(point 1 - point 2)**2 = {np.sum((np.array(point1) - np.array(point2))**2)}')
print(f'sqrt(sum((point 1 - point 2)**2)) = {np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))}')
```

## KNN Function

```python
def knn_predict(training_data, training_labels, test_point, k):
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
    
    print(f'Distances : {distances}')
    distances.sort(key=lambda x: x[0])
    print(f'Distances Sort : {distances}')
    
    k_nearest_labels = [label for _, label in distances[:k]]
    print(f'K Nearest Labels : {k_nearest_labels}')
    print(f'Counter(k_nearest_labels): {Counter(k_nearest_labels)}')
    print(f'Counter(k_nearest_labels).most_commont(1): {Counter(k_nearest_labels).most_common(1)}')
    return Counter(k_nearest_labels).most_common(1)[0][0]
```

## KNN Implementation

```python
training_data = [[1, 2], [2, 3], [3, 4], [6, 7], [7, 8]]
training_labels = ['A', 'A', 'A', 'B', 'B']
test_point = [4, 5]
k = 3
```

```python
prediction = knn_predict(training_data, training_labels, test_point, k)
print(f'Prediction: {prediction}')
```
